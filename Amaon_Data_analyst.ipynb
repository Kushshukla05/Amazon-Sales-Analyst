```python
import pandas as pd

# Sample DataFrame
data = {
    'Roll No': ['A,AB,AC', 'B,BC,BD', 'AB,A,AC'],
    'Product': ['kush', 'lksa', 'kas']
}

# Create DataFrame
df = pd.DataFrame(data)

# Function to normalize roll numbers (sort them)
def normalize_roll_no(roll_no):
    # Split the roll numbers by comma, sort them alphabetically, and join back to a string
    sorted_rolls = ",".join(sorted(roll_no.split(',')))
    print(sorted_rolls)
    return sorted_rolls

# Apply the function to normalize the 'Roll No' column
df['Normalized Roll No'] = df['Roll No'].apply(normalize_roll_no)

# Check for duplicates based on the normalized roll numbers
duplicates = df[df.duplicated('Normalized Roll No', keep=False)]

print("DataFrame with Normalized Roll No:")
print(df)

print("\nDuplicate rows based on Roll No patterns:")
print(duplicates)
```
